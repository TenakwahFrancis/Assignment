```python

name = input("Enter your Full name: ")
age = int(input("Enter your Age: "))
gpa = float(input("Enter your GPA (0.0 - 4.0): "))
courses_input = input("Enter your Courses (comma-separated): ")
courses = [course.strip() for course in courses_input.split(",")]

student_profile = {
    "name": name,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": len(courses)
}


deans_list = gpa >= 3.5
is_minor = age < 18

print("\n==== STUDENT PROFILE ====")
print(f"Name: {student_profile['name']}")
print(f"Age: {student_profile['age']}")
print(f"GPA: {student_profile['gpa']}")
print(f"Courses: {', '.join(student_profile['courses'])}")
print(f"Number of Courses: {student_profile['num_courses']}")
print(f"Dean's List: {deans_list}")


if is_minor:
    print("Status: Minor")
else:
    print("Status: Adult")

print("\n==== DATA TYPES ====")
print(f"Name = {type(name)}")
print(f"Age = {type(age)}")
print(f"GPA = {type(gpa)}")
print(f"Courses = {type(courses)}")
print(f"Dean's List = {type(deans_list)}")
print(f"Student Profile = {type(student_profile)}")

print("\nComplete Assignment 1 by implementing the student profile system.")
```
